```python
#Rock-Paper Scissor
p1 = [ ('r','s'),('p','r'),('s','p')]
import random
choice = [ 'r','p','s']
com = random.choice(choice)
player = input("Enter Your Choice (r,p,s) : ").strip()
ch = (player,com)
print("Computer's Choice : ",com)
print("User's Choice : ",player)
if ch in p1 : 
    print("Congrats you have won this game.")
elif com == player  :
    print("Match Tie ")
else :
    print("You such a Looser")
```
