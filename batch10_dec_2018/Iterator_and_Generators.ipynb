```python
class odd_iterator : 
    def __init__(self):
        self.current = 1
    def __iter__(self):
        self.end = 100
        return self
    def __next__(self):
        r = self.current 
        if self.current >= self.end :
            raise StopIteration("Number exceeds its limit")
        self.current += 2
        return r
```

```python
one = iter(odd_iterator())
```

```python
next(one)
```

```python
next(one)
```

```python
while True :
    print(next(one))
```

```python
class even_iterator : 
    def __init__(self):
        self.current = 0
    def __iter__(self):
        return self
    def __next__(self):
        self.current += 2
        return self.current
```

```python
inf = iter(even_iterator())
c = 1 
while c <= 100 : 
    print(next(inf),end=',  ')
    c = c + 1
```

```python
#generator

def even(s=2,n=100):
    while True : 
    
        if s % 2 : 
            s = s + 1
        yield s
        if s >= n : 
            raise StopIteration('Exceed limit')

        s = s + 2
    
```

```python
class Prime:
    def __init__(self,start):
        self.start = start
    def __iter__(self):
        return self
    def prime(self):
            num = self.start
            sq = round(sqrt(num)+1)
            for var in range(2,sq):
                if num % var == 0 :
                    self.start += 1
                    return self.prime()
            else :
                self.start += 1
                return num
    def __next__(self):
        return self.prime()

from math import sqrt
x = Prime(100000)
x = iter(x)
c = 0
for var in x :
    print(var,end=', ')
    c = c+1
    if c == 1000 : 
        break
print("Total numbers = ",c)
```

```python
class Hello:
    def __init__(self,max=0):
        self.max = max
    def __iter__(self):
        self.n = 0 
        return self
    def __next__(self):
        self.n += 1
        if self.n <= self.max :
            return self.n
        else :
            raise StopIteration("Limit Exceeds")

h = Hello(10)
h = iter(h)
while True :
    print(next(h))
```

```python
def even(num,max_limit):
    c = 0
    while c <= max_limit :
        if num % 2 == 0 :
            yield num
            num = num + 1
            c = c + 1
        else :
            num = num + 1
            c = c + 1
    
x = even(2,1000)
#for var in x :
#    print(var)
print(*x)
```
